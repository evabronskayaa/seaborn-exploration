# Задание 4. Bar Charts (гистограммы)

Допустим, мы хотим создать столбчатую диаграмму, показывающую среднюю задержку прибытия рейсов Spirit Airlines (код авиакомпании: NK) по месяцам.

```python
# Устанавливает высоту и ширину графика
plt.figure(figsize=(10,6))

# Добавляет заголовок для графика
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")

# Гистограмма, показывающая среднюю задержку прибытия рейсов Spirit Airlines по месяцам
sns.barplot(x=flight_data.index, y=flight_data['NK'])

# Подписывает ось y
plt.ylabel("Arrival delay (in minutes)")
```

### Вы можете сами выбрать фон для вашего графика
Это делается с помощью функции set_theme() (https://seaborn.pydata.org/generated/seaborn.set_theme.html)

```python
plt.figure(figsize=(10,6))
sns.set_theme(style="darkgrid")
sns.barplot(x=flight_data.index, y=flight_data['NK'])
```

```python
plt.figure(figsize=(10,6))
sns.set_theme(style="whitegrid")
sns.barplot(x=flight_data.index, y=flight_data['NK'])
```

### Вы можете сами выбрать палетку для вашего графика

Это делается с помощью функции diverging_palette() (https://seaborn.pydata.org/generated/seaborn.diverging_palette.html)

```python
plt.figure(figsize=(10,6))
palette = sns.diverging_palette(240, 10, n=9)
sns.barplot(x=flight_data.index, y=flight_data['NK'], palette = palette)
```

```python
sns.palplot(sns.diverging_palette(220, 20, n=11))
```

```python
sns.palplot(sns.diverging_palette(150, 275, s=80, l=55, n=11))
```

```python
sns.palplot(sns.diverging_palette(250, 15, s=75, l=40, n=11, center="dark"))
```

# Задание 5. Heatmaps (тепловые карты)

Еще один тип визуализации, о котором мы узнаем: тепловые карты!

В приведенной ниже ячейке кода мы создаем тепловую карту, чтобы быстро визуализировать шаблоны в файле flight_data.csv. Каждая ячейка имеет цветовую кодировку в соответствии с соответствующим значением.

```python
# Устанавливает высоту и ширину графика
plt.figure(figsize=(10,6))

# Добавляет заголовок для графика
plt.title("Average Arrival Delay for Each Airline, by Month")

# Тепловая карта, показывающая среднюю задержку прибытия рейсов Spirit Airlines по месяцам
sns.heatmap(data=flight_data, annot=True)

# Подписывает ось y
plt.xlabel("Airline")
```

Чтобы поменять палетку для нашей тепловой карты, нужно использовать атрибут cmap в функции heatmap(). Делается это следующим образом:

```python
plt.figure(figsize=(10,6))
cmap = sns.diverging_palette(250, 15, s=75, l=40, n=9, center="dark")
sns.heatmap(data=flight_data, cmap = cmap, annot=True)
```
